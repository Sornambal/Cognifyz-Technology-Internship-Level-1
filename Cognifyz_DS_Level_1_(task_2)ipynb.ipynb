```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("/content/Dataset .csv")
```

```python
# 1. Basic statistical measures for numerical columns
print("üìä Basic Statistical Summary:\n")
print(df.describe())
```

```python
# 2. Median for numerical columns
print("\nüìè Median of Numerical Columns:\n")
print(df.median(numeric_only=True))
```

```python
# 3. Standard deviation for numerical columns
print("\nüìê Standard Deviation of Numerical Columns:\n")
print(df.std(numeric_only=True))
```

```python
# 4. Distribution of categorical variables
print("\nüåç Country Code Distribution:\n")
print(df["Country Code"].value_counts().head(10))
print("\nüèôÔ∏è Top 10 Cities by Restaurant Count:\n")
print(df["City"].value_counts().head(10))

print("\nüçΩÔ∏è Top 10 Cuisines by Occurrence:\n")
print(df["Cuisines"].value_counts().head(10))
```

```python

# 5.Visualizations
plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='Price range', palette='pastel')
plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()

plt.figure(figsize=(14, 6))
top_cities = df['City'].value_counts().head(10)
sns.barplot(x=top_cities.index, y=top_cities.values, palette="Set2")
plt.xticks(rotation=45)
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Restaurant Count")
plt.tight_layout()
plt.show()
```
